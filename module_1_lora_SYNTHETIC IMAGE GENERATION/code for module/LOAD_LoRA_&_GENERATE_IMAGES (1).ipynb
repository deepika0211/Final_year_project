```python
import torch
import os
from diffusers import StableDiffusionPipeline
from peft import PeftModel
```

```python
BASE_MODEL = "runwayml/stable-diffusion-v1-5"
LORA_PATH  = "/content/drive/MyDrive/Final_year_project/lora_glioma"
OUTPUT_DIR = "/content/drive/MyDrive/Final_year_project/nf1_synthetic_images"

os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
pipe = StableDiffusionPipeline.from_pretrained(
    BASE_MODEL,
    dtype=torch.float16,
    safety_checker=None
).to("cuda")

pipe.unet.requires_grad_(False)
pipe.text_encoder.eval()
pipe.vae.eval()
```

```python
PROMPT = (
    "axial brain MRI slice, T1 weighted, clinical radiology scan, "
    "full brain cross section, visible ventricles, cortex and white matter, "
    "soft grayscale, low contrast, smooth intensity transitions, "
    "neurofibromatosis type 1, subtle internal nerve tissue abnormalities, "
    "hospital MRI machine output, diagnostic imaging"
)
```

```python
NEGATIVE_PROMPT = (
    "microscopy, histology, pathology slide, cellular texture, "
    "granular surface, speckled noise, tissue close-up, "
    "electron microscope, SEM, optical microscope, "
    "edge enhancement, thresholding, high contrast, "
    "3d render, illustration, text, watermark, color image"
)
```

```python
image = pipe(
    prompt=PROMPT,
    negative_prompt=NEGATIVE_PROMPT,
    num_inference_steps=60,   # ↑ more denoising = smoother
    guidance_scale=3.5,       # ↓ VERY IMPORTANT
    height=512,
    width=512
).images[0]

image.save(f"{OUTPUT_DIR}/nf3_test.png")
print("✅ NF1 image generated successfully")
```

```python
NUM_IMAGES = 200  # change as needed

for i in range(NUM_IMAGES):
    image = pipe(
        prompt=PROMPT,
        negative_prompt=NEGATIVE_PROMPT,
        num_inference_steps=30,
        guidance_scale=7.5,
        height=512,
        width=512
    ).images[0]

    image.save(f"{OUTPUT_DIR}/nf1_{i:04d}.png")

print("🎉 NF1 synthetic MRI dataset generated")
```
